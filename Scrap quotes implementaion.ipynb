```python
# Import the neccessary libraries.
import requests
from bs4 import BeautifulSoup
import csv
```

```python
# Send an HTTP request to the target webpage and fetch its raw HTML content.
url = "https://www.passiton.com/inspirational-quotes"
response = requests.get(url)
response.text
```

```python
# parse HTML using BeautifulSoup so we can easily navigate and extract specific parts of the content.
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Prepare a list to store all quote data which store each quote with (theme, image_url, lines, auther).
quotes = []
quote_boxes = soup.find_all('div', class_='text-center mb-8')

for box in quote_boxes:
    quote = {
        'theme': box.h5.text.strip() if box.h5 else None,
        'image_url': box.img['src'] if box.img else None,
        'lines': box.img['alt'].strip() if box.img and 'alt' in box.img.attrs else None,
        'author': box.find('p', class_='text-white/50').text.strip() 
                  if box.find('p', class_='text-white/50') 
                  else 'Unknown'
    }
    quotes.append(quote)

# Display some extracted quotes
for q in quotes[:5]:
    print(q)
```

```python
# Save extracted Data to CSV file named as "quotes".
filename = "quotes.csv"
with open(filename, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=['theme', 'image_url', 'lines', 'author'])
    writer.writeheader()
    for quote in quotes:
        writer.writerow(quote)
```
